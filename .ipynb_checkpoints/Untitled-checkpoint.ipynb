```python
HPDData = pd.read_csv('Historical Product Demand.csv')
```

```python
#Adding column to change date to number of days since the first day recorded
```

```python
datetime(year=int(1999), month=int(5), day=int(12))
```

```python
def createNewDateRow(row):
    try:
        return datetime(year=int(row['Date'].split("/")[0]), month=int(row['Date'].split("/")[1]), day=int(row['Date'].split("/")[2]))
    except:
        return np.nan
```

```python
HPDData['DateT'] = HPDData.apply(lambda row : createNewDateRow(row), axis = 1)
```

```python
HPDData
```

```python
HPDData = HPDData.dropna()
```

```python
HPDData
```

```python
minDate =  HPDData["DateT"].min()
HPDData['DateAsInt'] = HPDData.apply(lambda row : (row['DateT'] - minDate).days, axis = 1)
```
